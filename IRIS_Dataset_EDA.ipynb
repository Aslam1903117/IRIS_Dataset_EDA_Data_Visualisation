```python
# Import section
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
data=pd.read_csv('Iris.csv')
```

# changing  columns value to amke it  easy to understand


```python
data['Species']=data['Species'].replace("Iris-setosa","setosa")
data['Species']=data['Species'].replace("Iris-virginica","virginica")
data['Species']=data['Species'].replace("Iris-versicolor","versicolor")

```

```python
data
```

#changing name of the columns for easy to use

```python
data.columns=data.columns.str.lower()
```

```python
data
```

```python
#drop the id columns
data=data.drop(columns="id")
```

```python
data
```

#scatter plot with multiple layers

```python
plt.plot(data.sepallengthcm, data.sepalwidthcm,ls=' ' , marker ='o', label='sepal')
plt.plot(data.petallengthcm, data.petalwidthcm,ls=' ' , marker ='o', label='petal')
plt.legend()
```

# Ploting histogram using matplotlib

```python
data.hist(bins=50,figsize=(10,8))
plt.suptitle('Histogrsam For Iris Features')
plt.show
```

#Ploting histogram using Seaborn

```python
for features in data.columns:
    plt.figure(figsize=(6,4))
    sns.histplot(data[features],bins=50,kde=True)
    plt.title(f'Distribution of {features}')
    plt.xlabel(features)
    plt.ylabel('frequency')
    plt.show()
```

# pandas dataframe aproches

```python
data.groupby("species").mean().plot(color=['red','blue','black','green'],fontsize=10.0,figsize=(10,6))
```

```python
#pair plot for features
sns.pairplot(data, hue='species', diag_kind='kde', markers=["o", "s", "D"])
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)
plt.show()
```
